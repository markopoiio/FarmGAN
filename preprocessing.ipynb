```python
import pandas as pd
```

```python
df = pd.read_csv("drugs_phases234.csv", sep=";")
```

```python
df = df[["Smiles", "Indication Class"]]
# df.dropna(how="all", inplace=True)
df = df[df.Smiles.notna()]
df
```

```python
with open("smiles.smi", "w") as f:
    for smi in df.Smiles:
        f.write(smi + "\n")
    f.flush()
```

```python
from collections import defaultdict
```

```python
ind_classes = defaultdict(list)
for i, row in df.iterrows():
    for ind_class1 in str(row["Indication Class"]).split(";"):
        for ind_class in ind_class1.split(","):
            ind_class = ind_class.split(" (")[0]
            ind_class = ind_class.strip().lower()
            ind_classes[ind_class].append(row.Smiles)
```

```python
for k,v in ind_classes.items():
    print(k, len(v))
```
